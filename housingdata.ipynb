```python
#price prediction for housing price

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
data=pd.read_csv('Housing.csv')
```

```python
data.head()
data.tail()
data.info()
data.describe()
```

```python
data.columns
```

```python
plt.figure()
plt.scatter(data['price'],data['bedrooms'],color = 'blue',marker='+')
plt.xlabel("price")
plt.ylabel("bedrooms")
plt.show()
```

```python
sns.histplot(data['price'],kde=True,color='red')
```

```python
plt.figure(figsize=(8,8))
plt.scatter(data['bedrooms'],data['price'],label='bedrroms')
plt.scatter(data['floors'],data['price'],label='floors')
plt.xlabel("Number of bedrroms")
plt.ylabel("House Price")
plt.title("Number of bedrooms VS Price")
plt.legend()
plt.show()
```

```python
data.isna().sum()
```

```python
data["price"].fillna(data["price"].mean(), inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
numerical_data = data.select_dtypes(include=['number'])
```

```python
numerical_data.corr()
```

```python
plt.figure(figsize=(8,8))
plt.title("Correlation Matrix")
sns.heatmap(numerical_data.corr(), annot=True, fmt='.1f', annot_kws={'size':8})
```

```python
# visualizing target variable price
plt.figure(figsize=(8,8))
plt.title("Distribution of price")

sns.histplot(numerical_data['price'])
```

```python
# visualizing relation between sqft and price

sns.scatterplot(x=numerical_data['sqft_living'], y=numerical_data['price'])
plt.show()
```

simple linear regression

```python
x=numerical_data[['sqft_living']]
y=numerical_data['price']

print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=123)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

```python
linear.coef_
```

```python
linear.intercept_
```

```python
y_pred = linear.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
